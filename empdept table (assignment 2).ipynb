```python
# Join two tables
query = """
SELECT e.empno,
       e.ename,
       e.job,
       e.sal,
       d.dname,
       d.loc
FROM emp e
JOIN dept d
ON e.deptno = d.deptno
"""
```

```python
empdept = pd.read_sql(query, engine1)
```

```python
empdept
```

```python
# 1. Names of analysts and salesmen
empdept[
    empdept['job'].isin(['ANALYST', 'SALESMAN'])
]
```

```python
# Only names
empdept[
    empdept['job'].isin(['ANALYST', 'SALESMAN'])
]['ename']
```

```python
# Join two tables with all the coloumns
query = """
SELECT e.empno,
       e.ename,
       e.job,
       e.mgr,
       e.hiredate,
       e.comm,
       e.sal,
       e.deptno,
       d.dname,
       d.loc
FROM emp e
JOIN dept d
ON e.deptno = d.deptno
"""

empdept = pd.read_sql(query, engine1)

empdept
```

```python
# 2. Details of employees who have joined before 30 September 1981
empdept['hiredate'] = pd.to_datetime(empdept['hiredate'])

empdept[
    empdept['hiredate'] < '1981-09-30'
]
```

```python
# 3. Names of employees who are not managers
empdept[
    empdept['job'] != 'MANAGER'
]
```

```python
# Only names
empdept[
    empdept['job'] != 'MANAGER'
]['ename']
```

```python
# 4. Names of employees with specific employee numbers like 7369, 7521, 7839, 7934, 7788
empdept[
    empdept['empno'].isin([7369, 7521, 7839, 7934, 7788])
]
```

```python
# Only name and employee numbers
empdept[
    empdept['empno'].isin([7369, 7521, 7839, 7934, 7788])
][['ename','empno']]
```

```python
# 5. Employees not in 30, 40, and 10
empdept[
    ~empdept['deptno'].isin([30, 40, 10])
]
```

```python
# 6. Employees joined between 30-Jun-1981 and 31-Dec-1981
empdept[
    (empdept['hiredate'] >= '1981-06-30') &
    (empdept['hiredate'] <= '1981-12-31')
]
```

```python
# 7. Different designation in the company
empdept['job'].unique()
```

```python
# 8. Employees who are not eligible for commission
empdept[
    empdept['comm'].isnull()
]
```

```python
# 9. Employee who does not report to anybody
    empdept[
    empdept['mgr'].isnull()
][['ename', 'job']]
```

```python
# 10. Employees not assigned to any department
empdept[
    empdept['deptno'].isnull()
]
```

```python
# 11. Employees eligible for commission
empdept[
    empdept['comm'].notnull()
]
```

```python
# 12. Names starting or ending with S
empdept[
    empdept['ename'].str.startswith('S') |
    empdept['ename'].str.endswith('S')
]
```

```python
# 13. Names with the second character "i"
empdept[
    empdept['ename'].str[1] == 'I'
]
```

```python
# 14. Number of employees
empdept['empno'].count()
```

```python
# 15. Number of designations
empdept['empno'].count()
```

```python
# 16. Total salaries
empdept['sal'].sum()
```

```python
# 17. Maximum, minimum, and average salary
print("Maximum Salary:", empdept['sal'].max())
print("Minimum Salary:", empdept['sal'].min())
print("Average Salary:", empdept['sal'].mean())
```
