```python
from sqlalchemy import create_engine, text
import pymysql
import pandas as pd
import os
```

```python
engine1 = create_engine(
    "mysql+pymysql://root:piit2025@localhost:3306/piit"
)
print (engine1)
```

```python
# Function to run the query with parameters or query
def q(sql, **params):
    with engine1.connect() as conn:
        return pd.read_sql(text(sql), conn, params=params)
```

```python
q("SHOW DATABASES")
```

```python
q("SHOW TABLES")
```

```python
q("SELECT * FROM emp")
```

```python
# Count total employees
q("SELECT COUNT(*) FROM emp")
```

```python
q("SELECT COUNT(*) AS total_emp FROM emp")
```

```python
# Show employees by department
q("SELECT deptno, COUNT(*) AS total_emp FROM emp GROUP BY deptno")
```

```python
# Highest paying employee
q("""SELECT * 
    FROM emp 
        ORDER BY sal DESC 
            LIMIT 1""")
```

```python
# Top five highest paying employees
q("SELECT * FROM emp ORDER BY sal DESC LIMIT 5")
```

```python
# Lowest paying employee
q("SELECT * FROM emp ORDER BY sal ASC limit 1")
```

```python
# Lowest five paying employees
q("SELECT * FROM emp ORDER BY sal ASC LIMIT 5")
```

```python
# Average salary by department
q("SELECT deptno, AVG(SAL) AS avg_salary FROM emp GROUP BY deptno")
```

```python
# Employees getting commission
q("""SELECT ename, sal, comm
    FROM emp 
        WHERE comm IS NOT NULL""")
```

```python
# Employees aren't getting commission
q("SELECT ename, sal, comm FROM emp WHERE comm IS NULL")
```

```python
# Job-wise employee count
q("SELECT job, COUNT(*) AS total_emp FROM emp GROUP BY job")
```

```python
# employee getting more than average salary
q("""SELECT ename, job, sal 
    FROM emp 
        WHERE sal > (SELECT AVG(sal) FROM emp)""")
```

```python
# Department-wise salary report
q("""SELECT deptno,
  COUNT(*) AS total_emp,
  SUM(sal) AS total_salary,
  AVG(sal) AS avg_salary,
  MAX(sal) AS max_salary,
  MIN(sal) AS min_salary
  FROM emp
  GROUP BY deptno""")
```
