```python
from sqlalchemy import create_engine, text
import pymysql
import pandas as pd
import os
```

```python
engine1 = create_engine(
    "mysql+pymysql://root:piit2025@localhost:3306/highschool"
)
print (engine1)
```

```python
# Function to run the query with parameters or query
def q(sql, **params):
    with engine1.connect() as conn:
        return pd.read_sql(text(sql), conn, params=params)
```

```python
q("SHOW DATABASES")
```

```python
q("SHOW TABLES")
```

```python
q("SELECT * FROM highschool")
```

```python
# Average math score
q("SELECT AVG(math_score) AS avg_math FROM highschool")
```

```python
# Highest math-scoring student
q("""SELECT student_name, math_score
    FROM highschool
    ORDER BY math_score DESC
    LIMIT 1""")
```

```python
# Top 5 math-scoring student
q("""SELECT student_name, gender, grade, age, math_score
    FROM highschool
    ORDER BY math_score DESC
    LIMIT 5""")
```

```python
# Attendance analysis
q("""SELECT student_name, gender, grade, age, attendance_pct
    FROM highschool
    ORDER BY attendance_pct DESC""")
```

```python
# Average score by gender
q("""SELECT gender,
    AVG(math_score) AS avg_math,
    AVG(science_score) AS avg_science,
    AVG(english_score) AS avg_english
    FROM highschool
    GROUP BY gender""")
```

```python
# Students with low attendance
q("""SELECT student_name, attendance_pct
    FROM highschool
    WHERE attendance_pct <80""")
```

```python
# Students with more than 90 in math
q("""SELECT student_id, student_name, math_score, science_score, english_score
    FROM highschool
    WHERE math_score > 90""")
```

```python
# Students with more than 88 in English
q("""SELECT student_name, english_score, math_score, science_score
    FROM highschool
    WHERE english_score > 88""")
```

```python
# Students with more than 88 in English in descending order
q("""SELECT student_name, english_score, math_score, science_score
    FROM highschool
    WHERE english_score > 88
    ORDER BY english_score DESC""")
```

```python
# Graduation status report
q("""SELECT graduation_status,
    COUNT(*) AS total_students
    FROM highschool
    GROUP BY graduation_status""")
```

```python
# Top 5 students on overall scores
q("""SELECT student_name,
    (math_score + science_score + english_score) AS total_score
    FROM highschool
    ORDER BY total_score DESC
    LIMIT 5""")
```
