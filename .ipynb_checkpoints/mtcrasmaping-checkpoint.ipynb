```python
import pandas as pd
```

```python
from pydataset import data
mtcars = data('mtcars')
mtcars.head()
```

```python
mtcars['am'] = mtcars['am'].map({0:'auto', 1:'manual'})
mtcars.head()
```

```python
mtcars['vs'] = mtcars['vs'].map({0:'vshaped', 1:'straight'})
mtcars.head()
```

```python
mtcars['mpg'].max()
```

```python
mtcars['mpg'].min()
```

```python
mtcars['vs'] = mtcars['vs'].map({1:'vshaped', 2:'straight'})
mtcars.head()
```

```python
mtcars.pivot_table(values = 'mpg', index='gear', observed=False)
```

```python
mtcars.pivot_table(values = 'hp', index='cyl', observed=False)
```

```python
mtcars.pivot_table(values = ['mpg','hp'], index=['gear','am'], observed=True, 
                   aggfunc=['sum','mean','count']).reset_index()
```

```python
data = [
    [1, 'Taheera', 70, 86],
    [2, 'Aelita', 56, 90],
    [3, 'Khulan', 87, 90],
    [4, 'Batgerel', 71, 79],
    [5, 'Manisha', 88, 76],
]

df = pd.DataFrame(data, columns=['Rollno','Name', 'Maths', 'English'])
```

```python
df
```

wide to long

```python
df_long = df.melt(id_vars=['Rollno','Name'], 
                 value_vars=['Maths','English'], 
                 var_name='Subject', 
                 value_name='Marks')
df_long
```

```python
wide_df = df_long.pivot(index=['Rollno','Name'], columns='Subject',
             values='Marks').reset_index(drop=False).rename_axis(None, axis=1)
wide_df
```
